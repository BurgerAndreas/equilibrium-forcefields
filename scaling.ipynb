```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

sns.set_style()
```

```python
# plot different functions
l = 50
x = np.linspace(-l, l, 200)

# quadratic function
y1 = x**2
plt.plot(x, y1, label="y = x^2")

# power of 8 function
y5 = x**8
plt.plot(x, y5, label="y = x^8")

# exponential function
y2 = np.exp(x)
plt.plot(x, y2, label="y = e^x")

# N!
x = np.arange(1, l, dtype=int).tolist()
y3 = [math.factorial(_x) for _x in x]
plt.plot(x, y3, label="y = N!")

# plot
# log y scale
plt.yscale("log")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# logarithmic function
y3 = np.log(x)
plt.plot(x, y3, label="y = log(x)")

# sqrt function
y4 = np.sqrt(x)
plt.plot(x, y4, label="y = sqrt(x)")

# plot
plt.legend()
plt.tight_layout()
plt.show()
```
